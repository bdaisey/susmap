```python
import pandas as pd

ft = pd.read_csv("fairtrade30.csv")
ft.columns = ["name","addy1","addy2","city","state","zipcode","lat","lon","phone","website","industry"]
ft.head()
```

```python
cut = ft[["name","addy1","lat","lon","phone","website","industry"]]

cut.head()
cut.to_csv('ft30cut.csv')
```

from stackoverflow: https://stackoverflow.com/questions/48586647/python-script-to-convert-csv-to-geojson

```python
# convert cutdown .csv to json

import csv, json
from geojson import Feature, FeatureCollection, Point

features = []

with open('ft30cut.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader, None) # skip the headers
    for ftid, name, addy1, lat, lon, phone, website, industry in reader:
        lat,lon = map(float, (lat, lon))
        features.append(
            Feature(
                geometry = Point((lon,lat)),
                properties = {
                    'ftid': ftid,
                    'name': name,
                    'addy1': addy1,
                    'phone': phone,
                    'website': website,
                    'industry': industry
                }
            )
        )
        
collection = FeatureCollection(features)

with open("fairtrade30.geojson", "w") as f:
    f.write('%s' % collection)
        
```
