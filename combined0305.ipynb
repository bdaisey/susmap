```python
import pandas as pd

ft = pd.read_csv("combined0305.csv")
ft.columns = ["susid", "category", "name", "addy1", "addy2", "city", "state", "zipcode", "lat", "lon", "phone", "website", "industry"]
ft.head()
```

```python
# convert .csv to json

import csv, json
from geojson import Feature, FeatureCollection, Point

features = []

with open('combined0305.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader, None) # skip the headers
    for susid, category, name, addy1, addy2, city, state, zipcode, lat, lon, phone, website, industry in reader:
        lat,lon = map(float, (lat, lon))
        features.append(
            Feature(
                geometry = Point((lon,lat)),
                properties = {
                    'susid': susid,
                    'category': category,
                    'name': name,
                    'addy1': addy1,
                    'addy2': addy2,
                    'city': city,
                    'state': state,
                    'zipcode': zipcode,
                    'lat': lat,
                    'lon': lon,
                    'phone': phone,
                    'website': website,
                    'industry': industry
                }
            )
        )
        
collection = FeatureCollection(features)

with open("combined0305.geojson", "w") as f:
    f.write('%s' % collection)
        
print("Wrote geojson file")
```
